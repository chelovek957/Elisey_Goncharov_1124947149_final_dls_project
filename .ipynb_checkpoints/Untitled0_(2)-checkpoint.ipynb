```python
from torch.utils.data import random_split

train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_dataset, val_dataset, test_dataset = random_split(
    dataset, [train_size, val_size, test_size]
)
```

```python
from torch_geometric.loader import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=32)
test_loader  = DataLoader(test_dataset, batch_size=len(test_dataset))
```

```python
import torch
import torch.nn as nn

class BaselineMLP(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )

    def forward(self, x, batch):
        # x: [total_atoms, in_dim]
        # batch: индекс молекулы для каждого атома
        x_mean = torch.zeros(batch.max()+1, x.size(1), device=x.device)
        x_mean = x_mean.index_add(0, batch, x)
        counts = torch.bincount(batch)
        x_mean = x_mean / counts.unsqueeze(1)

        return self.net(x_mean).squeeze()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = BaselineMLP(dataset.num_features).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()
```

```python
def train_epoch(loader, model):
    model.train()
    total_loss = 0

    for batch in loader:
        batch = batch.to(device)


        optimizer.zero_grad()
        pred = model(batch.x, batch.batch)
        target = batch.y[:, 4]
        loss = loss_fn(pred, target)

        loss.backward()
        optimizer.step()

        total_loss += loss.item() * batch.num_graphs

    return total_loss / len(loader.dataset)
```

```python
def val_epoch(loader, model):
    model.eval()
    total_loss = 0
    total_mae = 0
    total_rmse = 0

    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)

            pred = model(batch.x, batch.batch)
            target = batch.y[:, 4]

            loss = loss_fn(pred, target)

            mae = torch.mean(torch.abs(pred - target))
            rmse = torch.sqrt(torch.mean((pred - target) ** 2))

            total_loss += loss.item() * batch.num_graphs
            total_mae += mae.item() * batch.num_graphs
            total_rmse += rmse.item() * batch.num_graphs

    n = len(loader.dataset)
    return total_loss / n, total_mae / n, total_rmse / n
```

```python
def test_epoch(loader,model):
  model.eval()
  total_loss=0
  with torch.no_grad():
    for batch in loader:
      batch=batch.to(device)
      pred=model(batch.x, batch.batch)
      target=batch.y[:,4]
      mae = torch.mean(torch.abs(pred - target))
      rmse = torch.sqrt(torch.mean((pred - target) ** 2))
      print("MAE:", mae)
      print("RMSE", rmse)

  return pred, target
```

```python
from tqdm import tqdm


min_delta = 1e-3
patience = 15


tr_hist = []
val_hist = []
mae_hist = []
rmse_hist = []


best_val = float("inf")
best_epoch = 0
wait = 0
best_state = None


num_epochs = 150
for epoch in tqdm(range(num_epochs), desc="Epochs"):


    tr_loss = train_epoch(train_loader, model)
    tr_hist.append(tr_loss)


    val_loss, val_mae, val_rmse = val_epoch(val_loader, model)
    val_hist.append(val_loss)
    mae_hist.append(val_mae)
    rmse_hist.append(val_rmse)


    print(
        f"Epoch {epoch+1:03d} | "
        f"Train Loss: {tr_loss:.4f} | "
        f"Val Loss: {val_loss:.4f} | "
        f"MAE: {val_mae:.4f} | "
        f"RMSE: {val_rmse:.4f}"
    )


    if val_mae < best_val - min_delta:
        best_val = val_mae
        best_epoch = epoch
        wait = 0
        # Save the model state on CPU
        best_state = {k: v.cpu() for k, v in model.state_dict().items()}
    else:
        wait += 1

    if wait >= patience:
        print(f"\nEarly stopping at epoch {epoch+1} (best MAE {best_val:.4f} at epoch {best_epoch+1})")
        break


print(f"Training finished. Best validation MAE: {best_val:.4f} at epoch {best_epoch+1}")
```

```python
from torch_geometric.nn import MessagePassing
import torch
import torch.nn as nn

class DistanceMP(MessagePassing):
    def __init__(self, hidden_dim):
        super().__init__(aggr="add")

        self.mlp = nn.Sequential(
            nn.Linear(hidden_dim + 1, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim)
        )

    def forward(self, x, pos, edge_index):
        return self.propagate(edge_index, x=x, pos=pos)

    def message(self, x_j, pos_i, pos_j):
        dist = torch.norm(pos_i - pos_j, dim=1, keepdim=True)
        msg = torch.cat([x_j, dist], dim=1)
        return self.mlp(msg)
```

```python
from torch_geometric.nn import global_mean_pool

class BaselineGNN(nn.Module):
    def __init__(self, in_dim, hidden_dim):
        super().__init__()

        self.embed = nn.Linear(in_dim, hidden_dim)

        self.conv1 = DistanceMP(hidden_dim)
        self.conv2 = DistanceMP(hidden_dim)

        self.out = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, batch):
        x, pos, edge_index, batch_idx = (
            batch.x,
            batch.pos,
            batch.edge_index,
            batch.batch,
        )

        x = self.embed(x)
        x = self.conv1(x, pos, edge_index)
        x = self.conv2(x, pos, edge_index)

        mol_emb = global_mean_pool(x, batch_idx)
        return self.out(mol_emb).squeeze(-1)
```

```python
import py3Dmol
import torch

def plot_molecule_3d(mol):
    """
    pos: (N, 3) atomic positions (torch or numpy)
    z: (N,) atomic numbers
    """
    pos=mol.pos
    z=mol.z
    if isinstance(pos, torch.Tensor):
        pos = pos.cpu().numpy()
    if isinstance(z, torch.Tensor):
        z = z.cpu().numpy()

    atomic_symbols = {
        1: "H", 6: "C", 7: "N", 8: "O", 9: "F"
    }

    xyz = f"{len(pos)}\n\n"
    for p, a in zip(pos, z):
        xyz += f"{atomic_symbols[int(a)]} {p[0]} {p[1]} {p[2]}\n"

    view = py3Dmol.view(width=400, height=400)
    view.addModel(xyz, "xyz")
    view.setStyle({"stick": {}, "sphere": {"scale": 0.3}})
    view.zoomTo()
    return view
```

```python
from torch_geometric.utils import scatter
import torch
import torch.nn as nn
from torch_geometric.nn import global_mean_pool
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class EGNNlayer(nn.Module):
  def __init__(self, in_dim, hidden_dim):
        super().__init__()

        self.MLP_pos = nn.Sequential(
            nn.Linear(in_dim*2+1, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1))

        self.MLP_feat = nn.Sequential(
            nn.Linear(in_dim*2+1, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim))

  def forward(self, x, pos, edge_index):

        res=edge_index[0]
        sen=edge_index[1]
        r=pos[res]-pos[sen]
        d=r.norm(dim=1, keepdim=True)
        feat_s=x[sen]
        feat_r=x[res]
        edge_input=torch.cat([feat_s, feat_r, d], dim=1)
        m=self.MLP_pos(edge_input)
        upd=scatter(m*r, res, dim=0, dim_size=x.size(0), reduce="add")
        pos=pos+upd
        edge_features=torch.cat([x[res], x[sen], d], dim=1)
        h=self.MLP_feat(edge_features)
        dx=scatter(h, res, dim=0, dim_size=x.size(0), reduce="add")
        x+=dx

        return x,pos
```

```python
class EGNNModel(nn.Module):
    def __init__(self, in_dim, hidden_dim, num_layers):
        super().__init__()
        self.embed = nn.Linear(in_dim, hidden_dim)
        self.layers = nn.ModuleList([
    EGNNlayer(hidden_dim, hidden_dim)
    for _ in range(num_layers)
])



        self.out = nn.Linear(hidden_dim, 1)

    def forward(self, batch):
        x, pos, edge_index, batch_idx = (
            batch.x,
            batch.pos,
            batch.edge_index,
            batch.batch,
        )
        x=self.embed(x)

        for layer in self.layers:
            x, pos = layer(x, pos, edge_index)

        mol_emb = global_mean_pool(x, batch_idx)
        return self.out(mol_emb).squeeze(-1)
```

```python
from ripser import ripser
import numpy as np
```

```python
def summarize_diag(beta):
  fin_mask=np.isfinite(beta[:,1])
  beta=beta[fin_mask]

  if beta.shape[0]==0:
    return 0, 0.0 ,0.0
  else:
    beta = beta[:,1]-beta[:,0]
    return beta.shape[0], beta.mean(), beta.max()
```

```python
def extract_tda(pos):
  diag=ripser(pos.cpu().numpy(), maxdim=2)['dgms']
  tda_feat=[]
  for i in range(0,2):
    stats=summarize_diag(diag[i])
    tda_feat.append(stats[0])
    tda_feat.append(stats[1])
    tda_feat.append(stats[2])
  return tda_feat

```

```python
def batch_extract_tda(batch):
    tda_feats = []
    for g in range(batch.num_graphs):
        mask = batch.batch == g
        pos_g = batch.pos[mask]
        tda_feats.append(extract_tda(pos_g))
    return torch.tensor(tda_feats, device=batch.pos.device, dtype=torch.float)
```

```python
class TDA_EGNNModel(nn.Module):
    def __init__(self, in_dim, hidden_dim, num_layers):
        super().__init__()
        self.embed = nn.Linear(in_dim, hidden_dim)
        self.layers = nn.ModuleList([
    EGNNlayer(hidden_dim, hidden_dim)
    for _ in range(num_layers)
])



        self.out = nn.Linear(hidden_dim, 1)

        self.graph_mlp = nn.Sequential(
    nn.Linear(hidden_dim + 6, hidden_dim),
    nn.ReLU(),
    nn.Linear(hidden_dim, hidden_dim)
)


    def forward(self, batch, tda_feat):
        x, pos, edge_index, batch_idx = (
            batch.x,
            batch.pos,
            batch.edge_index,
            batch.batch,
        )
        x=self.embed(x)

        for layer in self.layers:
            x, pos = layer(x, pos, edge_index)

        mol_emb = global_mean_pool(x, batch_idx)
        final_emb=torch.cat([mol_emb, tda_feat],dim=1)
        joint = torch.cat([mol_emb, tda_feat], dim=1)
        joint = self.graph_mlp(joint)


        return self.out(joint).squeeze(-1)
```

```python
max_id = max(idx_map) + 1
id_to_idx_tensor = torch.full(
    (max_id,), -1, dtype=torch.long
)

for i, qm9_id in enumerate(idx_map):
    id_to_idx_tensor[qm9_id] = i

id_to_idx_tensor = id_to_idx_tensor.to(device)
```

```python
def train_epoch(loader, model):
    model.train()
    total_loss = 0

    for batch in loader:
        batch = batch.to(device)

        qm9_ids = batch.idx.long()          # shape [B]
        idxs = id_to_idx_tensor[qm9_ids]    # GPU lookup
        extracts = tda_all_torch[idxs]      # GPU indexing

        optimizer.zero_grad()
        pred = model(batch, extracts)
        target = batch.y[:, 4]
        loss = loss_fn(pred, target)

        loss.backward()
        optimizer.step()

        total_loss += loss.item() * batch.num_graphs

    return total_loss / len(loader.dataset)
```

```python
train_epoch(train_loader, model4)
```

```python
def val_epoch(loader, model):
    model.eval()
    total_loss = 0
    total_mae = 0
    total_rmse = 0

    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)

            qm9_ids = batch.idx.long()           # [B]
            idxs = id_to_idx_tensor[qm9_ids]     # GPU lookup
            extracts = tda_all_torch[idxs]       # GPU indexing

            pred = model(batch, extracts)
            target = batch.y[:, 4]

            loss = loss_fn(pred, target)

            mae = torch.mean(torch.abs(pred - target))
            rmse = torch.sqrt(torch.mean((pred - target) ** 2))

            total_loss += loss.item() * batch.num_graphs
            total_mae += mae.item() * batch.num_graphs
            total_rmse += rmse.item() * batch.num_graphs

    n = len(loader.dataset)
    return total_loss / n, total_mae / n, total_rmse / n
```

```python
val_epoch(val_loader, model4)
```

```python
from tqdm import tqdm


min_delta = 1e-3
patience = 15


tr_hist = []
val_hist = []
mae_hist = []
rmse_hist = []


best_val = float("inf")
best_epoch = 0
wait = 0
best_state = None


num_epochs = 150
for epoch in tqdm(range(num_epochs), desc="Epochs"):


    tr_loss = train_epoch(train_loader, model4)
    tr_hist.append(tr_loss)


    val_loss, val_mae, val_rmse = val_epoch(val_loader, model4)
    val_hist.append(val_loss)
    mae_hist.append(val_mae)
    rmse_hist.append(val_rmse)


    print(
        f"Epoch {epoch+1:03d} | "
        f"Train Loss: {tr_loss:.4f} | "
        f"Val Loss: {val_loss:.4f} | "
        f"MAE: {val_mae:.4f} | "
        f"RMSE: {val_rmse:.4f}"
    )


    if val_mae < best_val - min_delta:
        best_val = val_mae
        best_epoch = epoch
        wait = 0
        # Save the model state on CPU
        best_state = {k: v.cpu() for k, v in model4.state_dict().items()}
    else:
        wait += 1

    if wait >= patience:
        print(f"\nEarly stopping at epoch {epoch+1} (best MAE {best_val:.4f} at epoch {best_epoch+1})")
        break


print(f"Training finished. Best validation MAE: {best_val:.4f} at epoch {best_epoch+1}")
```

```python
def plot_persistence_diagram(pos, maxdim=2):
    diag = ripser(pos, maxdim=maxdim)["dgms"]

    plt.figure(figsize=(5, 5))

    max_vals = []

    for i, dgm in enumerate(diag):
        if dgm.size == 0:
            continue

        finite = np.isfinite(dgm[:, 1])
        if finite.sum() == 0:
            continue

        plt.scatter(dgm[finite, 0], dgm[finite, 1], label=f"H{i}")
        max_vals.append(dgm[finite, 1].max())

    if len(max_vals) > 0:
        lim = max(max_vals)
        plt.plot([0, lim], [0, lim], "k--", alpha=0.5)

    plt.xlabel("Birth")
    plt.ylabel("Death")
    plt.legend()
    plt.title("Persistence Diagram")
    plt.show()
```

```python
def plot_barcodes(pos, dim=1):
    diag = ripser(pos, maxdim=dim)["dgms"][dim]
    finite = np.isfinite(diag[:, 1])
    diag = diag[finite]

    plt.figure(figsize=(6, 3))
    for i, (b, d) in enumerate(diag):
        plt.plot([b, d], [i, i], lw=2)

    plt.xlabel("Filtration value")
    plt.ylabel("Feature index")
    plt.title(f"Persistence Barcodes (H{dim})")
    plt.show()
```

```python
import py3Dmol
import torch

def plot_molecule_3d(pos, z):
    """
    pos: (N, 3) atomic positions (torch or numpy)
    z: (N,) atomic numbers
    """
    if isinstance(pos, torch.Tensor):
        pos = pos.cpu().numpy()
    if isinstance(z, torch.Tensor):
        z = z.cpu().numpy()

    atomic_symbols = {
        1: "H", 6: "C", 7: "N", 8: "O", 9: "F"
    }

    xyz = f"{len(pos)}\n\n"
    for p, a in zip(pos, z):
        xyz += f"{atomic_symbols[int(a)]} {p[0]} {p[1]} {p[2]}\n"

    view = py3Dmol.view(width=400, height=400)
    view.addModel(xyz, "xyz")
    view.setStyle({"stick": {}, "sphere": {"scale": 0.3}})
    view.zoomTo()
    return view
```
